```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd 
```

```python
def printImgs(images, n=1, m=1, title=''):
    
    images = images[:n*m]
    fig, axs = plt.subplots(
        nrows=n, ncols=m, 
        figsize=(15, 20), 
        subplot_kw={
            'xticks': [], 
            'yticks': []
        }
    )
    for ax, image in zip(axs.flat, images):
        ax.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.savefig(title)    
    plt.show()


def printGrafs(images, n=1, m=1, title=''):
    
    images = images[:n*m]
    fig, axs = plt.subplots(
        nrows=n, ncols=m, 
        figsize=(18, 20), 
        subplot_kw={
            'xticks': [], 
            'yticks': []
        }
    )
    for ax, image in zip(axs.flat, images):
        ax.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.savefig(title)    
    plt.show()
```

```python
lsNumPrue = ['390','392B','400B','400C']
```

```python
#BestImg340p1

lsBestImg = []

for Nprueb in lsNumPrue:
    ls = []
    for i in range(1,4):
        namefig = './' + Nprueb + '/BestImg' + Nprueb + 'p' + str(i) + '.png'
        imgOut = cv2.imread(namefig)
        ls.append(imgOut)
        
    lsBestImg.append(ls)
```
